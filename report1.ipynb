```python
def matrixout(mx, size):
  blanks = size * 3

  print("\u250c", end = "")
  for i in range(blanks):
    print("\u2500", end = "")
  print("\u2510")

  for i in range(size):
    for j in range(size):
      print("%1.3f" % mx[i][j], end = " ")
    print("")

  print("\u2514", end = "")
  for i in range(blanks):
    print("\u2500", end = "")
  print("\u2518")
```

```python
def transposeMatrix(m):
  return [[m[j][i] for j in range (len(m))] for i in range(len(m[0]))]
```

```python
def getMatrixMinor(m, i, j):
  return [row[:j] + row[j+1:] for row in (m[:i] + m[i+1:])]
```

```python
def getMatrixDeterminant(m):
  if len(m) == 1:
    return m[0][0]
    if len(m) == 2:
      return m[0][0]*[1][1] - m[0][1]*m[1][0]

  determinant = 0
  for c in range(len(m)):
    determinant += ((-1) ** c) * m[0][c] * getMatrixDeterminant(getMatrixMinor(m, 0, c))
  return determinant
```

```python
def has_inverse(m):
  determinant = getMatrixDeterminant(m)
  if abs(determinant) < 1e-10:
    print('오류 : 해당 행렬의 행렬식이 0이므로 역행렬 계산이 불가능합니다.')
    return False
  return True
```

```python
def getMatrixinverse(m):
  determinant = getMatrixDeterminant(m)

  if len(m) == 1:
    return [1.0/ m[0][0]]

  cofactors = []
  for r in range(len(m)):
    cofactorRow = []
    for c in range(len(m)):
      minor = getMatrixMinor(m, r, c)
      cofactorRow.append(((-1) ** (r + c)) * getMatrixDeterminant(minor))
    cofactors.append(cofactorRow)

  adjugate = transposeMatrix(cofactors)

  for r in range(len(adjugate)):
    for c in range(len(adjugate)):
      adjugate[r][c] = adjugate[r][c] / determinant
  return adjugate
```

```python
def Gauss_Jordan_inverse(m):
    n = len(m)

    I = [[1 if i == j else 0 for j in range(n)] for i in range(n)]
    augmented = [list(x + y) for x, y in zip(m, I)]

    for i in range(n):
        if augmented[i][i] == 0:
            for j in range(i+1, n):
                if augmented[j][i] != 0:
                    augmented[i], augmented[j] = augmented[j], augmented[i]
                    break

        pivot = augmented[i][i]
        augmented[i] = [x / pivot for x in augmented[i]]

        for j in range(n):
            if j != i:
                ratio = augmented[j][i]
                augmented[j] = [augmented[j][k] - ratio * augmented[i][k] for k in range(2 * n)]

    return [row[n:] for row in augmented]
```

```python
def matrix_equal_check(A, B):

    for i in range(len(A)):
        for j in range(len(A)):
            if abs(A[i][j] - B[i][j]) > 1e-10:
                return False
    return True
```

```python
def main():
  try:
    k = int(input("정방행렬의 차수를 입력하세요: "))
    if k <= 0:
      print("양의 정수를 입력해주세요.")
      return
    matrix_a = []
    for i in range(k):
      while True:
        try:
          row_input = input(f"{i+1}행: ").strip()
          row_values = [int(x) for x in row_input.split()]
          if len(row_values) != k:
            print(f"입력 오류: 정확히 {k}개의 값을 입력해야 합니다.")
            continue
          matrix_a.append(row_values)
          break
        except ValueError:
           print("입력 오류: 숫자만 입력해주세요.")

    if has_inverse(matrix_a):
      print("행렬식으로 구한 역행렬:")
      inverse_a = getMatrixinverse(matrix_a)
      matrixout(inverse_a, k)

      print("가우스-조던 소거법으로 구한 역행렬:")
      inverse_b = Gauss_Jordan_inverse(matrix_a)
      matrixout(inverse_b, k)

      if(matrix_equal_check(inverse_a, inverse_b)):
        print("두 방법의 결과값이 일치합니다.")
      else:
        print("두 방법의 결과값이 일치하지 않습니다.")

  except ValueError:
    print("입력 오류: 숫자만 입력해주세요.")



```
